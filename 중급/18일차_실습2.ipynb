```python
from surprise.dataset import Dataset
```

```python
data = Dataset.load_builtin('ml-1m')
```

```python
data.construct_testset([('4916','231',1.0,4)])
```

```python
dir(data)
```

```python
from surprise.model_selection import train_test_split
```

```python
train, test = train_test_split(data)
```

```python
from surprise.prediction_algorithms import KNNBaseline, KNNBasic, SVD
```

```python
knn1 = KNNBasic()
```

```python
knn1.fit(train)
```

```python
knn1.predict('1','1')
```

```python
knn1.predict('1','1',5)
```

```python
knn1.predict('1','2')
```

```python
t = knn1.test(test)
```

```python
from surprise import accuracy
```

```python
accuracy.mse(t)
```

```python
accuracy.rmse(t)
```

```python
knn2 = KNNBaseline()
```

```python
knn2.fit(train)
```

```python
knn2.predict('1','1',5)
```

```python
tt = knn2.test(test)
```

```python
accuracy.mse(tt)
```

```python
accuracy.rmse(tt)
```
