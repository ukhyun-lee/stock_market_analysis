# supervised : dimension reduction => 선형 판별 분석
c.f. 자연어처리 에서 latent dirichiet allocation

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
Ida = LinearDiscriminantAnalysis(n_components=2)   #=> 클래스 -1 까지가 최대값 n_componets 정함 , y값으로 차원축소하는 지도학습
```

```python
Ida.fit(X,y).transform(X)
```

#차원 축소 비선형 방법=> 복잡하지만 성능이 좋을 수 있음

```python
from sklearn.manifold import TSNE
```

```python
ts = TSNE()
```

```python
XX = ts.fit_transform(X,y)
```

```python
import pandas as pd
```

```python
y
```

```python
data = pd.DataFrame(XX)
```

```python
data['target'] = y
```

```python
data.rename(columns={0:'x', 1:'y'}, inplace=True)
```

```python
import seaborn as sns
```

```python
sns.scatterplot(data, x='x', y='y', hue='target')
```
