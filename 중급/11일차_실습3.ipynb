## 차원축소 그래프 그리기

```python
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
```

```python
pca = PCA(n_components=2)
tsne = TSNE()
```

```python
data_pca = pca.fit_transform(data_sub.iloc[:,:-1])
data_pca_pd = pd.DataFrame(data_pca)
data_pca_pd['Class'] = data_sub.Class.values
data_pca_pd.rename(columns={0:'X', 1:'Y'}, inplace=True)
```

```python
data_tsne = tsne.fit_transform(data_sub.iloc[:,:-1], data_sub.Class)
data_tsne_pd = pd.DataFrame(data_tsne)
data_tsne_pd['Class'] = data_sub.Class.values
data_tsne_pd.rename(columns={0:'X', 1:'Y'}, inplace=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(data_sub.iloc[:,:-1], data_sub.Class)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.model_selection import cross_val_score
```

```python
#cross_val_score(LogisticRegression(C=1), X_train, y_train, cv=10)
cross_val_score(LogisticRegression(C=1), X_train, y_train, cv=10).mean()
```

```python
cross_val_score(LogisticRegression(C=10, solver='newton-cg'), X_train,y_train, cv=10).mean()
```

# hypeermeter tuning

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, lr.predict(X_test)))
```

```python
from sklearn.model_selection import learning_curve
```

```python
train_size, train_score, test_score = learning_curve(LogisticRegression(), X_train, y_train, cv=10)
```
