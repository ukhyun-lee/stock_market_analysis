# 결즉치 (missing data) = impute

```python
mpg = sns.load_dataset('mpg')
```

```python
mpg[mpg.horsepower.isna()]
```

```python
mpg_ = mpg[mpg.horsepower.isna()]
```

```python
mpg.fillna(method='ffill')  #fillna(3) forward fill backward fill
```

```python
mpg_.assign(horsepower=1) # ==> 중요
```

```python
mpg_.assign(horsepower=lambda x:x.cylinders)
```
