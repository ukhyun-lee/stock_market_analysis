```python
#5  나하고 가까운 애들이 안본 영화 빼야
```

```python

def find_unseen_movies_by_user(user_id, n=5):
    others = cor_similarity.loc[user_id].nlargest(n+1)[1:].index
    others_movies = data[data['UserID'].isin(others)].MovieID.values
    mine_movies = data[data['UserID']==user_id].MovieID.values
    movie_index = np.setdiff1d(others_movies, mine_movies)
    return data[data['MovieID'].isin(movie_index)]
```

```python
#6 나하고 가까운 애들이 안본 영화 중 rank 몇이상 추천
```

```python

def find_unseen_movies_by_user_rating(user_id, n=5, r=4):
    others = cor_similarity.loc[user_id].nlargest(n+1)[1:].index
    others_movies = data[data['UserID'].isin(others)].MovieID.values
    mine_movies = data[data['UserID']==user_id].MovieID.values
    movie_index = np.setdiff1d(others_movies, mine_movies)
    rating_index = movie_rating[(movie_rating['MovieID'].isin(movie_index)) & (movie_rating['Rating'] >= r)]
    return rating_index
```
