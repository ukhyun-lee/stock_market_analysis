```python
iqr = tips.tip.quantile(0.75) - tips.tip.quantile(0.25)
```

```python
tips.tip.quantile(0.25)
```

```python
outlier_max = 1.5 * iqr + tips.tip.quantile(0.75)
```

```python
outlier_min = tips.tip.quantile(0.25) - 1.5*iqr
```

```python
tips.tip > outlier_max
```

```python
outlier_min
```

```python
tips[tips.tip > outlier_max]
```
