```python
def check(x):
    if x['Average value'] <= x['Good(Blue)']:
        return 'Blue'
    elif x['Good(Blue)'] < x['Average value'] <= x['Normal(Green)']:
        return 'Green'
    elif x['Normal(Green)'] < x['Average value'] <= x['Bad(Yellow)']:
        return 'Yellow'
    elif x['Bad(Yellow)'] < x['Average value'] <= x['Very bad(Red)']:
        return 'Red'
```

```python
seoul['eval'] = seoul.apply(check, axis=1)
```

```python
seoul.drop(columns=['Good(Blue)','Normal(Green)','Bad(Yellow)','Very bad(Red)'], inplace=True)
```

```python
seoul.drop(columns=['base_year', 'Address', 'Unit of measurement'], inplace=True)
```

```python
seoul.drop(columns=['code','Station name(district)','Station code','Item code'], inplace=True)
```

geo

```python
seoul_geo = seoul[seoul['Instrument status'] == 0].copy()
```

```python
seoul_geo.drop(columns='Instrument status', inplace=True)
```

```python
seoul_geo.set_index('Measurement date', inplace=True)
```

```python
def advanced_filtering(item, time=None, location=None):
    temp = seoul_geo[seoul_geo['Item name'] == item]
    if location is not None:
        temp =  temp[temp['name'] == location]
    if time is not None:
        temp = temp.loc[time]
    return temp
```

```python
advanced_filtering('SO2', location='성북구')
```

```python
advanced_filtering('SO2', location='성북구')[['Average value']].plot.line()
```

```python
t = advanced_filtering('CO', time='2017-12-05 17')
```

```python
t.plot('Average value', legend=True)
```

```python
s = t.reset_index()
```

```python
s.drop(columns='Measurement date', inplace=True)
```
