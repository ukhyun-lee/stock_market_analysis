# EDA

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(iris.iloc[:,:-1], iris.target)
```

```python
from yellowbrick.features import Rank2D, Rank1D, RadViz, ParallelCoordinates
```

```python
ra = Rank2D()
ra.fit(X_train, y_train)
ra.transform(X_train)
ra.show()
```

```python
ra = Rank1D()
ra.fit(X_train, y_train)
ra.transform(X_train)
ra.show()
```

```python
ra = ParallelCoordinates()
ra.fit(X_train, y_train)
ra.transform(X_train)
ra.show()
```

```python
data = load_iris(as_frame=True)
iris = data.frame
X_train,X_test,y_train,y_test = train_test_split(iris.iloc[:,:-1], iris.target)
```

# pandas 지원 버그 > numpy로 변경

```python
ra = RadViz()
ra.fit(X_train.values, y_train.values)  ###numpy로
ra.transform(X_train.values)
ra.show()
```

# Classification

```python
from sklearn.linear_model import LogisticRegression
```

```python
from yellowbrick.classifier import ClassificationReport, ClassPredictionError, ConfusionMatrix, \
PrecisionRecallCurve, ROCAUC, DiscriminationThreshold
from sklearn.metrics import classification_report
```

```python
lr = LogisticRegression()
```

```python
report = ClassificationReport(lr)
report.fit(X_train,y_train)
report.score(X_test,y_test)
report.show()
```

```python
report = ClassPredictionError(lr)
report.fit(X_train,y_train)
report.score(X_test,y_test)
report.show()
```

```python
report = ConfusionMatrix(lr)
report.fit(X_train,y_train)
report.score(X_test,y_test)
report.show()
```

```python
report = PrecisionRecallCurve(lr)
report.fit(X_train,y_train)
report.score(X_test,y_test)
report.show()
```

```python
report = ROCAUC(lr)
report.fit(X_train,y_train)
report.score(X_test,y_test)
report.show()
```
