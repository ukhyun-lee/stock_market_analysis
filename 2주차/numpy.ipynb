정수 배열을 사용하여 원하는 인덱스의 값을 지정(put)하거나 가져올 수(take) 있음

```python
arr = np.arange(10)
index = [5, 3, 2, 6]

# index에 있는 값을 반환
arr.take(index)
```

##(참고) 기타 연산

where: 조건절

```python
arr = np.arange(10).reshape(2,5)

# arr의 원소가 5보다 크면 10, 그렇지 않은 경우에는 0을 반환
np.where(arr> 5, 10, 0)
```

sort: 정렬

```python
arr = np.random.randn(5)
print('정렬 이전:', arr)
arr.sort()
print('정렬 결과:', arr)
```

```python
arr = np.random.randn(9).reshape(3, 3)
print('정렬 이전:\n', arr)
arr.sort(axis = 1)
print('정렬 결과:\n', arr)
```

분위수: 팬시 색인 + 정렬

```python
arr = np.random.randn(100)
print(arr)
arr.sort()

quantile_1 = arr[int(0.25 * len(arr))]
quantile_4 = arr[int(0.75 * len(arr))]

print('\n\n25%:', quantile_1, '75%:', quantile_4)
```

# 3.1.6 저장

 - ndarray를 바이너리 형식으로 디스크에 저장할 수 있음
 - 대용량 데이터의 경우 읽고 쓰는데 시간을 줄일 수 있음

```python
arr = np.arange(10).reshape(2,5)
arr
```

```python
np.save('array', arr)
```

```python
load_arr = np.load('array.npy')
```

```python
load_arr
```

(참고)
np.savez()
- 여러 배열을 압축된 형식으로 저장할 수 있음
- 확장자: npz
- 저장하려는 배열을 키워드 인자 형태로 전달

```python
arr1 = np.arange(10).reshape(2,5)
arr2= np.arange(10)
np.savez('array.npz', a = arr1, b= arr2)
```

```python
arrs = np.load('array.npz')
arrs
```

```python
arrs['a']
```

```python
arrs['b']
```

```python
arrs.close()
```
