# 4.1.2 산점도
x축과 y축이 모두 실수형 변수
x축과 y축의 상관관계(경향성)


seaborn 사용방법
 - sns.플롯이름(x=x축에 설정할 컬럼명, y=y축에 설정할 컬럼명)

```python
sns.scatterplot(x=df['total_bill'], y=df['tip'])
```

```python
sns.scatterplot(x=df['sex'], y=df['tip'])
```

```python
# hue: 범주형 변수를 설정, 범주에 따라 다른 색상을 표시
sns.scatterplot(x=df['total_bill'], y=df['tip'], hue=df['sex']) #smoker
```

```python
sns.scatterplot(x=df['total_bill'], y=df['tip'], hue=df['sex'], size=df['size'])
```

-----
<div class="alert alert-warning">
[Tip] 표시할 데이터가 적을 때의 산점도와 막대그래프로 시각화
</div>

```python
x= [1, 2, 3, 4, 5]
y= [10, 20, 25, 30, 42]
sns.scatterplot(x=x, y=y)
```

```python
sns.barplot(x=x, y=y)
```

# 4.1.4 선그래프(lineplot)
- 시간의 흐름에 따른 데이터의 분포를 확인
- 예) 월별 매출 변화, 일별 주사 변화, 시간별 온도 변화

```python
x= [1, 2, 3, 4, 5]
y= [10, 20, 25, 30, 42]
sns.lineplot(x=x, y=y)
```

```python
sns.lineplot(x=df['size'], y=df['tip']) #ci 신뢰구간
```

-----
<div class="alert alert-warning">
[Tip] 신뢰구간의 변경
</div>

```python
print("min:{}, max:{}".format(min(df[df['size'] == 5]['tip']), max(df[df['size'] == 5]['tip'])))
```

```python
sns.lineplot(x=df['size'], y=df['tip'], ci=100)
```

```python
sns.lineplot(x=df['size'], y=df['tip'], ci=None)
```

----------

```python
sns.lineplot(x=df['size'], y=df['tip'], hue=df['day'],  ci=None)
```

-----
<div class="alert alert-warning">
[Tip] hue 속성은 범주형(category) 변수를 지정해주세요!
</div>

```python
# hue 속성에 여러 개의 값이 포함된 변수를 지정한 경우
sns.lineplot(x=df['day'], y=df['total_bill'], hue=df['tip'],  ci=None)
```
