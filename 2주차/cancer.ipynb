 ## 실습 문제
 최근 한 병원에서 종양을 정확하게 진단 할 수 있는 알고리즘을 사용하기 시작했습니다.  
 
 'Radius (worst)'값이 25인 종양이 주어질 경우,   
 * 종양을 Benign(양성)으로 분류할 가능성이 더 높습니까,   
 * 아니면 Malignant(악성)으로 분류 할 가능성이 더 높습니까?  

```python
sns.histplot(data=cancer, x = 'Radius (worst)', hue='Diagnosis')
```

종양이 있는지 여부를 판단하기 위해 사용할 수 있는 속성(feature)은 무엇이 있을까? 데이터 시각화를 통해 속성을 찾아보고, 설명하세요.

```python
cancer.columns
```

```python
nrows, ncols = 5, 6
fig, axs = plt.subplots(nrows=nrows, ncols=ncols)
fig.set_size_inches(20,20)

for i in range(nrows):
    for j in range(ncols):
        attr = i * ncols + j + 1
        sns.histplot(data=cancer, x=cancer.columns[attr],  hue='Diagnosis', kde=True, ax=axs[i][j])
```
